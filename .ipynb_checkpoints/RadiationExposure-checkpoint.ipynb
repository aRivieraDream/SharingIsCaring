```python
def f(x):
    import math
    return 10*math.e**(math.log(0.5)/5.27 * x)
```

```python
def radiationexposure(start, stop, step):
    if start >= stop:
        return float(0)
    else:
        print 'f(%r) = %r' % (start, f(start))
        return (f(start) * step) + radiationexposure(start + step, stop, step)
```

```python
test1 = radiationexposure(0, 5, 1)
print test1
```

```python
test2 = radiationexposure(5, 11, 1)
print test2
```
